```python
'''
Link to challenge
https://preppindata.blogspot.com/2023/01/2023-week-3-targets-for-dsb.html 
'''
import pandas as pd
```

```python
trans_df = pd.read_csv('PD 2023 Wk 1 Input.csv')
trans_df.head(5)
```

```python
target_df = pd.read_csv('Targets.csv')
target_df.head(5)
```

```python
#  filter the transactions to just look at DSB, These will be transactions that contain DSB in the Transaction Code field
trans_df.insert(0,'XX','')
trans_df['XX'] = [x.split('-')[0] for x in trans_df['Transaction Code']]
trans_df.head(5)
```

```python
trans_df.info()
```

```python
new_trans_df = pd.DataFrame(columns= trans_df.columns)
for i in range(len(trans_df.index)):
    if trans_df.iloc[i,0] == 'DSB':
        new_trans_df.loc[len(new_trans_df.index)] = trans_df.iloc[i,:]
new_trans_df.info()
new_trans_df.head(6)
```

```python
#  Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values
for i in range(len(new_trans_df.index)):
    x = new_trans_df.loc[i,'Online or In-Person']
    if x == 1:
        new_trans_df.loc[i,'Online or In-Person'] = 'Online'
    elif x == 2:
         new_trans_df.loc[i,'Online or In-Person'] = 'In-Person'

new_trans_df.head(5)
```

```python
#  Change the date to be the quarter
def date2quarter(x):
    d = x.split(' ')[0]
    m = d.split('/')[1]
    m = int(m)
    if m in [1,2,3]:
        return 1
    elif m in [4,5,6]:
        return 2
    elif m in [7,8,9]:
        return 3
    elif m in [10,11,12]:
        return 4

for i in range(len(new_trans_df.index)):
    new_trans_df.loc[i,'Quarter'] = date2quarter(new_trans_df.loc[i,'Transaction Date'])

new_trans_df.head(5)
```

```python
#  Sum the transaction values for each quarter and for each Type of Transaction (Online or In-Person)
trans_df = pd.DataFrame(new_trans_df.groupby(['Online or In-Person','Quarter']).Value.sum().reset_index())
trans_df['Quarter'] = trans_df['Quarter'].astype(int)
trans_df.head(5)
```

```python
target_df.head(5)
```

```python
tar_df = pd.melt(target_df, id_vars='Online or In-Person', var_name='Quarter', value_name='Target').reset_index(drop=True)
tar_df.head(10)
```

```python
tar_df['Quarter'] = [int(x[1]) for x in tar_df['Quarter']]
tar_df['Quarter'] = tar_df['Quarter'].astype(int)
tar_df.head(10)
```

```python

trans_df.head(10)
```

```python
# Join the two datasets together
joined_df = pd.merge(trans_df, tar_df, on=['Quarter', 'Online or In-Person'])
joined_df.head(10)
```

```python
#  Calculate the Variance to Target for each row 

joined_df['Variance to Target'] = [x-y for x,y in zip(joined_df['Value'], joined_df['Target'])]
joined_df.head(10)
```

```python
joined_df.to_csv('output.csv' , index=False)
```

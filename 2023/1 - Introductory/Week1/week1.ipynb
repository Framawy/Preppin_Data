```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('PD 2023 Wk 1 Input.csv')
df.head(5)
```

```python
df.insert(0, 'Bank', None)
df.head(5)
```

```python
#Split the Transaction Code to extract the letters at the start of the transaction code

def trans_code(x):
    s = x.split('-')
    bank = s.pop(0)
    s = '-'.join(s)
    return bank,s

for i in range(len(df['Transaction Code'])):
    x,y = trans_code(df.iloc[i]['Transaction Code'])
    if (i%100==0): print(x,y)
    df.loc[i,'Bank'] = x
    df.loc[i,'Transaction Code'] = y

df.head(5)
```

```python
#Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values

for i in range(len(df['Online or In-Person'])):
    x = df.loc[i, 'Online or In-Person']
    if x == 1 :
        df.loc[i, 'Online or In-Person'] = 'Online'
    elif x == 2:
        df.loc[i, 'Online or In-Person'] = 'In-Person'

df.head(5)
```

```python
type(df['Transaction Date'][0])
```

```python
#Change the date to be the day of the week 
def date2day(d):
    d = d.split(' ')[0]
    d,m,y = d.split('/')
    date = datetime.datetime(int(y),int(m),int(d)).strftime('%A')
    
    return date

for i in range(len(df['Transaction Date'])):
    df.loc[i,'Transaction Date'] = date2day(df.loc[i,'Transaction Date'])

df.head(5)
```

```python
#Output 1: Total Values of Transactions by each bank

bank_list = {}
for i in range(len(df['Bank'])):
     if df.loc[i,'Bank'] in bank_list:
         bank_list[df.loc[i,'Bank']] += df.loc[i,'Value']
     else:
         bank_list[df.loc[i,'Bank']] = df.loc[i,'Value']

output1 = pd.DataFrame.from_dict(bank_list , orient = 'index')
output1 = output1.reset_index()
output1.rename(columns={'index':'Bank', 0:'value'}, inplace=True)
output1.to_csv('output1.csv',index=False)
output1

```

```python
#Output 2: Total Values by Bank, Day of the Week and Type of Transaction
output2 = pd.DataFrame(df.groupby(['Bank','Online or In-Person','Transaction Date']).Value.sum().reset_index())
output2.to_csv('output2.csv',index=False)
output2.info()
```

```python
#Output 3: Total Values by Bank and Customer Code

output3 = pd.DataFrame(df.groupby(['Bank', 'Customer Code']).Value.sum().reset_index())
output3.to_csv('output3.csv',index=False)
output3.info()
```

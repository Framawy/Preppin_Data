```python
'''
Challenge link
https://preppindata.blogspot.com/2023/03/2023-week-13-rolling-stock-price-trends.html
'''

import pandas as pd
import numpy as np
import datetime as datetime
```

### Input all the months of data available


```python
df_list = []
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
for i in range(1,13):
    if i == 1:
        df_list.append(pd.read_csv('MOCK_DATA.csv'))
        df_list[i-1]['Month'] = months[i-1]
    else:
        df_list.append(pd.read_csv(f'MOCK_DATA-{i}.csv'))
        df_list[i-1]['Month'] = months[i-1]
print(len(df_list))
df_list[-1][:5]
```

```python
df_list[0][:5]
```

```python
df = pd.concat(df_list)
df.info()
```

```python
df = df.dropna(subset = ['Sector'])
df.info()
```

### Create a Trade Order field showing the order each trade has happened in for each Sector from the File Name and ID as the order of each trade within each monthly file


```python

df['Trade Order'] = df.groupby('Sector')['id'].rank(method='first')

df[:10]
```

### Remove all data fields except:
        Trade Order
        Sector
        Purchase Price

```python

df.drop(columns=['id', 'first_name', 'last_name', 'Ticker', 'Market', 'Stock Name', 'Market Cap', 'Month'], inplace=True)

df[:5]
```

```python
df.info()
```

### Create a Parameter to allow the user to select the rolling number of trades to be incorporated into the moving average. 
        I've set a default of 3 trades in my moving average

```python
avg = 3
```

### Create a data set that records the previous 2 trades (where applicable) as well as that Trade Order record.
### Workout the Rolling Average Purchase Price for each Trade Order in each Sector

```python
df = df.sort_values(['Sector','Trade Order']).reset_index()
df.drop(columns=['index'], inplace=True)
df[:10]
```

```python
df['Purchase Price'] = [x.replace('$', '') for x in df['Purchase Price']]
df['Purchase Price'] = df['Purchase Price'].astype(float)
df.info()
```

```python
sectors_df = df
sectors_df['Rolling AVG'] = 0
sectors_df['Rolling AVG'] = df.groupby('Sector')['Purchase Price'].rolling(3).mean().reset_index(0,drop=True)
sectors_df[:10]
```

### Filter the data for the last 100 trades for each Sector

```python
result_df = sectors_df.groupby('Sector').tail(100)
result_df[:10]
```

### Create the Previous Trades field to show the oldest trade (1) through to the latest trade (100).

```python
result_df['Previous Trades'] = result_df.groupby('Sector')['Trade Order'].rank('first')
result_df[:10]
```

```python
result_df.to_csv('output.csv', index=False)
```
